```python
from gensim.models import Word2Vec
from sklearn.cluster import KMeans
```

```python
model = Word2Vec.load("word2vec.model")
model1 = Word2Vec.load("FastText.model")
```

```python
x = model.wv.vectors
kmeans = KMeans(n_clusters=5, max_iter=300)
kmeans.fit(x)
```

```python
sorted(list(zip(kmeans.labels_, model.wv.vocab)))
```

```python
y = model1.wv.vectors
kmeans1 = KMeans(n_clusters=5, max_iter=300)
kmeans1.fit(y)
```

```python
sorted(list(zip(kmeans1.labels_, model1.wv.vocab)))
```
