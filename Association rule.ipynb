```python
from tqdm import tqdm
from bs4 import BeautifulSoup
import json
from nltk.corpus import stopwords
import jieba
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
path='constellation.json'
constellation_content=[]
with open(path, encoding = 'utf8') as f:
    for line in tqdm(f):
        try:
            j_content = json.loads(line)
            soup=BeautifulSoup(j_content['content'])
            constellation_content.append(soup.text)
        except:
            pass
```

```python
constellation_content[0:1]
```

```python
type(constellation_content)
```

```python
constellation_cut_content = []
for i in constellation_content:
    constellation_cut_content.append(list(jieba.cut(i)))
```

```python
for i in constellation_cut_content:
    for w in i:
        if len(str(w)) < 2:
            i.remove(w)
```

```python
stop_words1 =open("stopwords1.txt", 'r').read()
stop_words =open("stopwords.txt", 'r').read()
```

```python
for i in constellation_cut_content:
    for w in i:
        if w in stop_words:
            i.remove(w)
```

```python
for i in constellation_cut_content:
    for w in i:
        if w in stop_words1:
            i.remove(w)
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(constellation_cut_content).transform(constellation_cut_content)
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df.info()
```

```python
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
```

```python
rank =  frequent_itemsets[0:64]
```

```python
from mlxtend.frequent_patterns import association_rules

minconf = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```
